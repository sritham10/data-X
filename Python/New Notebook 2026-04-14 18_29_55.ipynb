```python
def count_in_list(arr,number):
    cnt=0
    for el in arr:
        if el==number:
            cnt+=1
    return cnt


arr =[1,2,4,5,1]
print(count_in_list(arr,1))
```

```python
from collections import deque
dq = deque([2,3,1])
dq.append(5)
print(dq)

dq.appendleft(1)
print(dq)

dq.pop()
print(dq)

dq.popleft()
print(dq)

dq.extend([1,2,3])
print(dq)

dq.extendleft([1,2,3])
print(dq)

dq.rotate(2)
print(dq)
```

```python
from collections import Counter
arr = Counter([1,2,3,4,5,6,7,5,4,8,9,10])
print((arr[5]))
print(arr.most_common(2))
print(arr.most_common())
```
