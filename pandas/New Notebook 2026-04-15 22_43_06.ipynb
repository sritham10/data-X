```python
import pandas as pd
df = pd.read_csv('employees.csv')
display(df)
```

```python
display(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull())
```

Handling null

```python
df_modified = df.dropna(subset=['salary'])
display(df_modified)
```

```python
# Fill Nulls with Defaults
df['department'] = df['department'].fillna('Unknown')
df['salary'] = df['salary'].fillna(df['salary'].median())
```

```python
display(df)
```

Convert and clean date column

```python
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')  # Convert invalid dates to NaT
df['join_date'] = df['join_date'].fillna(method='ffill')  # Fill NaT with forward fill
display(df)
```

Handling outliers

```python
print(df['salary'].describe())
```
